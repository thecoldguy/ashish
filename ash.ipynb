```python
import matplotlib.pyplot as plt

data=[50,79,15,79,5]
plt.bar(range(len(data)),data,color='g')
# plt.plot(data,marker='p',linestyle=':',color='r',linewidth=2)
plt.title("New Horizon",fontsize=40)
plt.xlabel("Python",fontsize=30)
plt.ylabel("Management",fontsize=30)
plt.grid(False)
plt.show()
```

```python
import matplotlib.pyplot as plt

price=[10,20,100,40,50]
demand=[70,80,60,40,30]
supply=[100,40,20,80,100]

plt.figure(figsize=(8, 6))
plt.plot(price, demand, label='Demand', marker='o',color='green')
plt.plot(price, supply, label='Supply', marker='o',color='red')

plt.title('New Suppy and Demand Curve')
plt.xlabel('Price')
plt.ylabel('Quantity')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

data=[20,80,23,30,5]

plt.figure(figsize=(5, 6))
plt.pie(data,labels=data, autopct='%1.2f%%')
plt.title('Pie Chart')
plt.show()
```
